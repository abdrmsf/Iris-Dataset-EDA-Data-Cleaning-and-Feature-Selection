## EDA

```python
plt.figure(figsize=(10,8))
numerical_columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
value=1
for cols in numerical_columns:
    plt.subplot(2,2,value)
    sns.histplot(x=df[cols],kde=True)
    value+=1

plt.tight_layout()
```

## From the above graph we can see that most flowers have average sepallength and sepalwidth(due to being normally distributed)

## But we can notice a gap in PetalLengthCm and PetalWidthCm ---> the left to the gap is Iris setosa and the right to the gap is Iris-Versicolor and Iris-Virginica as observed in the dataset ---> so these two columns are very important for the dataset

```python
plt.figure(figsize=(10,8))
numerical_columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
value=1
for cols in numerical_columns:
    plt.subplot(2,2,value)
    sns.boxplot(x=df[cols],hue=df['Species'])
    value+=1

plt.tight_layout()
```

## The Iris Setosa in both PetalLengthCm and PetalWidthCm are clearly seperated to the left while Versicolor and Virginica are in the right with less gap between them,these petal features help us distingush between Setosa and others.
## While in the Sepal Function there is overlapping between Versicolor and Virginica, cannot distinguish between them.Setosa is also not very clearly seperated
